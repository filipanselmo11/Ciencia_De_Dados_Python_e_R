## Algoritmo de treinamento do Neur√¥nio Perceptron

### Soma das entradas com seu respectivos pesos

```python
def calculo(x, w):
    u = 0
    for i in range(len(w)):
        u+= x[i] * w[i]
    return u
```

```python
pesos = [-0.5441, -0.5562, -0.4024]
#df = [-1,2,2]
result = calculo(df, pesos)
print(result)
```

```python
def funcao_ativacao(x, w) :
    if(calculo(x,w) >= 0) :
        return 1
    else:
        return 0
```

```python
def ajuste_pesos(e, wi, xi) :
    taxa_aprendizado = 0.1
    aux = taxa_aprendizado * e * xi
    w_ajustado = wi + aux
    return w_ajustado
```
