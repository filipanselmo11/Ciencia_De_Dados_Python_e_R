# Resolvendo um Problema Linearmente Separável

## Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Algoritmo de Treinamento do Perceptron

```python
def step_function(u):
    if(u >= 0):
        return 1
    return 0
```

```python
def perceptron(N, e, W):
    epoca = 1
    iscomplet = 0
    update = 0
    
    while(iscomplet != 1000):
        
        print("Época =========> ", epoca)
        print("Vetor de peos: ", W)
        aprendeu = 0
        errou = 0
        
        for i in range(1000):
            xi = X[i]
            
            yd = Y[i]
            
            u = np.dot(xi, W)
            
            y = step_function(u)
            
            e = yd - y
            
            W = W + N*e*xi
            
            if(not e):
                aprendeu += 1
            else:
                errou += 1
                update += 1
        print("Exemplos corretos: ", aprendeu)
        print("Exemplos incorretos: ", errou)
        
        iscomplet = aprendeu
        epoca += 1
        
        print("\n\n")
        
    return (W, epoca, update)
            
```

## Função para visualização dos gráficos

```python
def plotpredict(X, Y, W):
    color = ['^b', '^r']
    
    for k  in range(1000):
        entradas = X[k]
        plt.plot(entradas[1], entradas[2], color[Y[k]])
        
    x = np.random.uniform(-0.5, 0.5, 20)
    
    y = W[0]/W[2] - (W[1]/W[2]) * x
    plt.plot(x, y)
    plt.xlabel('Coordenada x1')
    plt.ylabel('Coordenada x2')
    plt.show()
```

```python
def plotgrafico(X, Y, W):
    color = ['^b', '^r']
    for k in range(1000):
        entradas = X[k]
        plt.plot(entradas[1], entradas[2], color[Y[k]])
    
    x = np.random.uniform(-0.5, 0.5, 10)
    
    y = (W[0]/W[2]) - (W[1]/W[2]) *  x
    plt.plot(x, y)
    plt.xlabel('Coordenada x1')
    plt.ylabel('Coordenada x2')
    plt.show()
```

## Entrada

```python
dt = np.dtype([('min', float), ('sec', float), ('temp', float)])
df = np.fromfile('dataAll.txt', dtype=dt)
yD = []
n_points = len(df)

termos = []

print("N termos:", n_points)
for i in range(n_points):
    termo=[]
    termo.append(-1)
    termo.append(round((df[i][0]), 4))
    termo.append(round((df[i][1]), 4))
    
    termos.append(termo)
    
for i in range(n_points):
    yD.append(int((df[i][2])))
    
for i in range(len(termos)):
    termos[i] = np.asarray(termos[i])
    
X = termos
Y = yD

n_points1 = len(X)
n_points2 = len(Y)
print("n termos X: ", n_points1)
print("n termos Y: ", n_points2)
```

```python
N = 0.1
e = 1
W = np.random.uniform(-0.5, 0.5, 3)
```

## Saída

```python
W, epoca, update = perceptron(N, e, W)
print("Quantidade de épocas até a convergência:", epoca-1)
print("Quantas vezes houve ajuste em W:", update)
```
