```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('./Dados_dashboards/trees.csv')
```

```python
h = np.histogram(base.iloc[:, 1], bins=2)
h
```

```python
#Histograma
plt.hist(base.iloc[:,1], bins=10)
plt.title("Arvores")
plt.ylabel("Frequencia")
plt.xlabel("Altura")
```

```python
#Gráfico de densidade
sns.distplot(base.iloc[:,1], hist=True, kde=True,
    bins=6, color="blue", hist_kws={'edgeColor': 'black'})
```

```python
#Gŕafico de dispersão
plt.scatter(base.Girth, base.Volume, color="blue",
    facecolors='none')#Ver documentação para mais detalhes
plt.title("Arvores")
plt.xlabel("Volume")
plt.ylabel("Girth")
```

```python
#Gráfico de Linha
plt.plot(base.Girth, base.Volume)
```

```python
sns.regplot(base.Girth, base.Volume, data=base,
    x_jitter=0.3, fit_reg=False)
```

```python
#Dispersão com legendas
base2 = pd.read_csv('./Dados_dashboards/co2.csv')
```

```python
x = base2.conc
y = base2.uptake

unicos = list(set(base2.Treatment))
unicos
```

```python
for i in range(len(unicos)):
    indice = base2.Treatment == unicos[i]
    plt.scatter(x[indice], y[indice], label=unicos[i])
plt.legend(loc='lower right')
```

```python
dados3 = pd.read_csv('./Dados_dashboards/co2.csv')
dados3
```

```python
sns.scatterplot(dados3.conc, dados3.uptake, hue=dados3.Type)
```

```python
#q = quebec
q = dados3.loc[dados3['Type'] == 'Quebec']
m = dados3.loc[dados3['Type'] == 'Mississippi']
```

```python
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(q.conc, q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
sns.scatterplot(m.conc, m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
ch = dados3.loc[dados3['Treatment'] == 'chilled']
nc = dados3.loc[dados3['Treatment'] == 'nonchilled']
```

```python
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(ch.conc, ch.uptake).set_title('Chilled')
plt.subplot(1,2,2)
sns.scatterplot(nc.conc, nc.uptake).set_title('Non Chilled')
plt.tight_layout()
```
