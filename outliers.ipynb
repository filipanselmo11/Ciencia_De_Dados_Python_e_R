```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.csv')
iris
```

```python
plt.boxplot(iris.iloc[:, 1])#Com outliers
```

```python
plt.boxplot(iris.iloc[:, 1], showfliers=False) #Sem outliers
```

```python
outliers = iris[(iris['sepal_width'] > 4.0) | (iris['sepal_width'] < 2.1)]
outliers
```

```python
from pyod.models.knn import KNN
```

```python
sepal_width = iris.iloc[:,1].values
```

```python
sepal_width = sepal_width.reshape(-1,1)
```

```python
detector = KNN()
detector.fit(sepal_width)
```

```python
previsoes = detector.labels_
previsoes
```
