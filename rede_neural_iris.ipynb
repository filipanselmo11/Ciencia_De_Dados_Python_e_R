```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from yellowbrick.classifier import ConfusionMatrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
data = datasets.load_iris()
data
```

```python
previsores = data.data
classe = data.target
```

```python
classe_dummy = np_utils.to_categorical(classe)
print(classe_dummy)
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(previsores,
                                                       classe_dummy,
                                                       test_size = 0.3,
                                                       random_state = 0)
```

```python
modelo = Sequential()
modelo.add(Dense(units=5, input_dim=4))
modelo.add(Dense(units=4))
modelo.add(Dense(units=3, activation='softmax'))
```

```python
modelo.summary()
```

```python
#ImpressÃ£o
from keras.utils.vis_utils import plot_model
```

```python
modelo.compile(optimizer='adam', loss='categorical_crossentropy',
              metrics=['accuracy'])
modelo.fit(X_treino, y_treino, epochs=1000,
          validation_data=(X_teste, y_teste))
```

```python
previsoes = modelo.predict(X_teste)
print(previsoes)
```

```python
previsoes = (previsoes > 0.5)
print(previsoes)
```

```python
y_teste_matrix = [np.argmax(t) for t in y_teste]
y_previsao_matrix = [np.argmax(t) for t in previsoes]
```

```python
confusao = confusion_matrix(y_teste_matrix, y_previsao_matrix)
print(confusao)
```
